The file provided in this exercise is a CSV file which stands for Comma-Separated Values. **üîé More about CSV files:**

A **CSV (Comma-Separated Values)** file is a simple way to store data in a table format. Each line of the file represents a row in the table, with values separated by commas. CSV files are widely used for storing time-series data, such as temperature readings taken at different times. 

For example, the table below is an example of how a CSV table would look like:

| Time  | Temperature (¬∞C) |
|-------|------------------|
| 10/19/06 | 17.8           |
| 10/20/06 | 18.1             |
| 10/21/06 | 18.3             |
| 10/22/06 | 19.1             |

**CSV** files are easy to read and can be opened by tools like Excel or processed by programming languages like Python.

````python
# Reformatting the time-series into a dataframe

temp["Date"] = pd.to_datetime(temp['t'], errors='coerce')#format = 'mixed')#'%Y-%m-%d')
temp.set_index("Date", inplace=True)
temp.drop(['t'], axis=1, inplace=True)
temp.rename(columns={'temp': 'Temp'}, inplace=True)

temp.head()
````

````python

temp.plot(figsize=(18, 4))
#plt.xlim('2006','2024')
````

````python
# Compute day-of-the-year climatology
data_clim_doy = temp.groupby(temp.index.dayofyear).mean()
data_clim_doy.plot(figsize=(10, 3));

# Compute day-of-the-year climatology on the same time index
data_ano_doy_ind = temp.groupby(pd.to_datetime(temp.index).dayofyear).transform('mean')
data_ano_doy_ind.plot(figsize=(10, 3));
plt.xlim('2015','2023')

# Compute anomalies
data_ano_doy = temp - data_ano_doy_ind
data_ano_doy.head()
data_ano_doy.plot(figsize=(10, 3));
plt.xlim('2014-03-01','2022-06-01')
````

````python
# Compute day-of-the-year climatology of the 90th percentile
data_clim_thres_doy = temp.groupby(temp.index.dayofyear).quantile([.9])
data_clim_thres_doy.plot(figsize=(10, 3));

# Compute day-of-the-year climatology of the 90th percentile on the same time index
data_clim_thres_doy_ind = temp.groupby(pd.to_datetime(temp.index).dayofyear).transform('quantile', .9)
data_clim_thres_doy_ind.plot(figsize=(10, 3));

# Compute anomalies from the 90th percentile
data_ano_thres_doy = temp - data_clim_thres_doy_ind
data_ano_thres_doy.head()
data_ano_thres_doy.plot(figsize=(10, 3));
````
